```python
import pandas as pd

data = pd.read_csv("autos.csv")
```

```python
data = data.dropna()
```

```python
price_array = data['price'].to_numpy()
```

```python
price_array
```

```python
years_array = data['yearOfRegistration'].to_numpy()
```

```python
edited_years_array = 2024-years_array
```

```python
edited_years_array
```

```python
import matplotlib.pyplot as plt

plt.hist(edited_years_array, bins=10, edgecolor = 'black')
plt.xlabel("Edited Years")
plt.ylabel("Frequency")
plt.title("Histogram of Edited Years")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(edited_years_array, price_array, c='blue' , alpha = 0.5)
plt.xlabel("Years")
plt.ylabel("Price")
plt.title("Relation between prices and years")
plt.show()
```

```python
import numpy as np
Q1 = np.percentile(price_array, 25)
Q3 = np.percentile(price_array, 75)
```

```python
IQR = Q3 - Q1
lower_bound = Q1 - 1.5* IQR
upper_bound = Q3 + 1.5* IQR
```

```python
edited_price_array = price_array[(price_array >= lower_bound) & (price_array <= upper_bound)]
```

```python
edited_price_array
```

```python
matching_years_array = data[data['price'].isin(edited_price_array)]['yearOfRegistration']
```

```python
matching_edited_years_series = 2024-matching_years_array
```

```python
matching_edited_years_array = matching_edited_years_series.to_numpy()
```

```python
matching_edited_years_array
```

```python
import matplotlib.pyplot as plt

plt.scatter(matching_edited_years_array, edited_price_array, c='blue' , alpha = 0.5)
plt.xlabel("Years")
plt.ylabel("Price")
plt.title("Relation between prices and years")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
X= matching_edited_years_array
y = edited_price_array
```

```python
X
```

```python
X1 =X.reshape(-1,1)
```

```python
X1
```

```python
y1 = y.reshape(-1,1)
```

```python
y1
```

```python
X_train, X_test, y_train, y_test = train_test_split(X1, y1, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y1_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y1_pred)
```

```python
mse
```

```python
plt.scatter(X_test, y_test, color='blue', label='Actual')
plt.plot(X_test, y1_pred, color='red', linewidth=2, label='Predicted')
plt.xlabel('Years')
plt.ylabel('Price')
plt.title('Linear Regression: Actual vs. Predicted')
plt.legend()
plt.show()
```
